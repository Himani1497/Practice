**<h2>ABOUT NETFLIX<h2>**

Netflix is one of the most popular media and video streaming platforms. They have over 10000 movies or tv shows available on their platform, as of mid-2021, they have over 222M Subscribers globally. This tabular dataset consists of listings of all the movies and tv shows available on Netflix, along with details such as - cast, directors, ratings, release year, duration, etc.

**<h2>Business Problem<h2>**

Analyze the data and generate insights that could help Netflix ijn deciding which type of shows/movies to produce and how they can grow the business in different countries

**<h2>Dataset<h2>**

The dataset provided to you consists of a list of all the TV shows/movies available on Netflix:


*   Show_id: Unique ID for every Movie / Tv Show
*   Type: Identifier - A Movie or TV Show
*   Title: Title of the Movie / Tv Show
*   Director: Director of the Movie
*  Cast: Actors involved in the movie/show
*   Country: Country where the movie/show was produced
*  Date_added: Date it was added on Netflix
*  Release_year: Actual Release year of the movie/show
*  Rating: TV Rating of the movie/show
*   Duration: Total Duration - in minutes or number of seasons
*   Listed_in: Genre
*   Director: Director of the Movie
*   Description: The summary description

**<u><h2>Handling null values<h2><u>**











```python
print('Total missing values accross all columns:')
df.isnull().sum().to_frame()
```

**<h3>Insights<h3>**It can be seen that the missing 'duration' column value are present in 'rating' column. lets shift the value from 'rating' column

```python
# fixing 'duration' column missing values
ind = df[df['duration'].isna()].index
df.loc[ind] = df.loc[ind].fillna(method = 'ffill' , axis = 1)
df.loc[ind ,'rating'] = 'Not Available'
df.loc[ind]
```

```python
df[df.rating.isna()]
```

```python
# fixing 'rating' column missing values
indices = df[df.rating.isna()].index
df.loc[indices , 'rating'] = 'Not Available'
df.loc[indices]
```

```python
df.rating.unique()
df.rating.value_counts()
```

**<h3>Insights<h3>**In rating column , NR (Not rated) is same as UR (Unrated). lets change UR to NR.

```python
df.loc[df['rating'] == 'UR' , 'rating'] = 'NR'
df.rating.value_counts()
```

```python
print('Total missing values accross all columns:')
df.isnull().sum().to_frame()
```

```python
# fixing 'director', 'cast', 'country', 'date_added' columns missing values
categorical_columns = ['director', 'cast', 'country', 'date_added']
for i in categorical_columns:
  df[i].fillna(f'Unknown {i.capitalize()}', inplace=True)
df.head()
```

```python
print('Total missing values accross all columns:')
df.isnull().sum().to_frame()
```

**<u><h2>Un-nesting the columns<h2><u>**



```python
# Un-nesting the 'cast' column
cast_df = df[["show_id", "cast"]]
cast_df["list_of_cast"] = cast_df["cast"].apply(lambda x: str(x).split(', '))
cast_df = cast_df.explode("list_of_cast")
cast_df.head(5)
```

```python
# Un-nesting the 'country' column
country_df = df[["show_id", "country"]]
country_df["list_of_country"] = country_df["country"].apply(lambda x: str(x).split(', '))
country_df = country_df.explode("list_of_country")
country_df.head()
```

```python
# Un-nesting the 'listed_in' (genre) column
genre_df = df[["show_id", "listed_in"]]
genre_df["list_of_genre"] = genre_df["listed_in"].apply(lambda x: str(x).split(', '))
genre_df = genre_df.explode("list_of_genre")
genre_df.head()
```

```python
# Un-nesting the 'director' column
director_df = df[["show_id", "director"]]
director_df["list_of_director"] = director_df["director"].apply(lambda x: str(x).split(', '))
director_df = director_df.explode("list_of_director")
director_df.head()
```

```python

tv_shows = df.query('type == "TV Show"')
movies = df.query('type == "Movie"')

# Counting the number of titles per week and finding the week with the highest count
tvdata  = tv_shows.groupby('week_added')['title'].count().sort_values(ascending=False).to_frame()
mvdata = movies.groupby('week_added')['title'].count().sort_values(ascending=False).to_frame()

#assigning variable to max count and week
max_week_tv = tvdata.index[0]
max_title_tv = tvdata.iloc[0]['title']
max_week_mv = mvdata.index[0]
max_title_mv = mvdata.iloc[0]['title']

#setting the plot style
fig = plt.figure(figsize = (15,10))
gs = fig.add_gridspec(2,1)
# creating graph for count of TV show
ax0 = fig.add_subplot(gs[0,:])
sns.lineplot(data = tvdata, x = "week_added", y = "title",marker = 'o',markersize = 10,color ='red',label = 'TV Show',ax = ax0)
ax0.annotate(f'week: {max_week_tv}', xy=(max_week_tv, max_title_tv), xytext=(max_week_tv, max_title_tv), fontsize=12)
#customizing axis labels
ax0.set_title('WoW data of TV show Launch',fontweight = 'bold')
plt.xlabel('Week')
plt.ylabel('Counts')
# creating graph for count of movies
ax1 = fig.add_subplot(gs[1,:])
sns.lineplot(data = mvdata, x = "week_added", y = "title",marker = 'o',markersize = 10,color ='green',label = 'Movies', ax = ax1)
ax1.annotate(f'week: {max_week_mv}', xy=(max_week_mv, max_title_mv), xytext=(max_week_mv, max_title_mv),fontsize=12)
#customizing axis labels
plt.xlabel('Week')
plt.ylabel('Counts')
ax1.set_title('WoW data of Movie Launch',fontweight = 'bold')

plt.show()
```

**<h3>Insights<h3>**
* The highest volume of movies are typically uploaded during the 1st week of the year.
*The highest volume of TV shows tends to be added in the 27th week of the year.
* A lot of new films are added during intial week, but less are added during the year. show that netflix upload the movie mostly altogether at the starting of the year.
* In the case of TV shows, there is no sudden spike in addition trend. instead, episode-by-episode content is uploaded every few weeks throughout the year.

**<h3>Recommendations<h3>**
* Netflix should concentrate more on negotiating contracts and deals with film makers.therefore as soon as the film is released, it is uploaded.Being able to compete better with other platforms can be aided by this.






```python
#converting month_added to categorical type to help in future sorting steps
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
tv_shows['month_added'] = pd.Categorical(tv_shows['month_added'], categories=months, ordered=True)
movies['month_added'] = pd.Categorical(movies['month_added'], categories=months, ordered=True)

# Counting the number of titles per month and finding the month with the highest count
tv_shows_monthly = tv_shows.groupby('month_added')['title'].count().sort_values(ascending=False).to_frame()
movies_monthly = movies.groupby('month_added')['title'].count().sort_values(ascending=False).to_frame()

#assigning variable to max count and month
max_month_tv = tv_shows_monthly.index[0]
max_mtitle_tv = tv_shows_monthly.iloc[0]['title']
max_month_mv = movies_monthly.index[0]
max_mtitle_mv = movies_monthly.iloc[0]['title']

#setting the plot style
fig = plt.figure(figsize = (13,8))
# creating graph for count of movies
ax0 = fig.add_subplot(gs[0,:])
sns.lineplot(data = tv_shows_monthly, x = "month_added", y = "title",marker = 'o',markersize = 10,color ='red',label = 'TV Show',ax = ax0)
ax0.annotate(f'month: {max_month_tv}', xy=(max_month_tv, max_mtitle_tv), xytext=(max_month_tv, max_mtitle_tv), fontsize=12)
#customizing axis labels
ax0.set_title('MoM data of TV show Launch',fontweight = 'bold')
plt.xlabel(None)
plt.ylabel('Counts')
# creating graph for count of tv shows
ax1 = fig.add_subplot(gs[1,:])
sns.lineplot(data = movies_monthly, x = "month_added", y = "title",marker = 'o',markersize = 10,color ='green',label = 'Movies', ax = ax1)
ax1.annotate(f'month: {max_month_mv}', xy=(max_month_mv, max_mtitle_mv), xytext=(max_month_mv, max_mtitle_mv),fontsize=12)
#customizing axis labels
plt.xlabel(None)
plt.ylabel('Counts')
ax1.set_title('MoM data of Movies Launch',fontweight = 'bold')

plt.show()
```

**<h3>Insights<h3>**
* The most number of movies are uploaded in july month.
*The most number of TV show are uploaded in december  month.
*  Least content is uploaded in month of february




**<h3>Recommendations<h3>**
*  Plan content releases strategically based on insights about the best months for TV shows and movies. Aligning releases with peak months or with holidays week in month can maximize viewership and engagement.

```python
df['Days_Between'] = (date_format_df['date_added'] - pd.to_datetime(df['release_year'], format='%Y')).dt.days
df.tail()
```

```python
mode_days = df['Days_Between'].mode()[0]
plt.figure(figsize=(15, 6))
plt.plot(df['Days_Between'].value_counts().sort_index(), color='orange', linestyle="-")
plt.axvline(x=mode_days, color='black', linestyle='--', label='Mode: {} days'.format(mode_days))
plt.title('Distribution of Days Between Release Year and Date Added')
plt.xlabel('Number of Days')
plt.ylabel('Shows Count')
plt.legend()
plt.grid(True)
plt.show()
```

**<h3>Insights<h3>**
* mode  of no. of days between release year and date added is 334 days. this thow there are high chances that the show/movie will uploaded 334th day after relased
*High number of content is added in between 0 to ~2000 days

**<h3>Recommendations<h3>**

For example, they can prioritize acquiring shows that tend to gain popularity quickly after release to minimize the time between release and availability on Netflix. films .
